```python
import pandas as pd
import os
from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt

# --- 1. Define Paths ---
DATASET_FOLDER = os.path.join(BASE_PATH, 'dataset/')
TRAIN_IMAGE_FOLDER = os.path.join(BASE_PATH, 'images/train/')

# --- 2. Load your Sample DataFrame ---
sample_csv_path = os.path.join(DATASET_FOLDER, 'train_sample.csv')
df_sample = pd.read_csv(sample_csv_path)

# --- 3. Helper Function to get Image Path ---
def get_image_path(image_link):
    filename = Path(image_link).name
    return os.path.join(TRAIN_IMAGE_FOLDER, filename)

# --- 4. Pick an Image to Display ---
# Let's get the image for the 100th product in your sample
product_row = df_sample.iloc[101]
image_url = product_row['image_link']
image_local_path = get_image_path(image_url)

print(f"Displaying image for product with sample_id: {product_row['sample_id']}")
print(f"Image path: {image_local_path}")

# --- 5. Open and Display the Image ---
try:
    # Open the image file using PIL
    img = Image.open(image_local_path)

    # Use matplotlib to display the image
    plt.figure(figsize=(6, 6))
    plt.imshow(img)
    plt.axis('off')  # Hide the axes
    plt.show()

except FileNotFoundError:
    print(f"\n‚ùå ERROR: Image not found at the specified path.")
    print("Please check if the images have finished downloading to that folder.")
```
